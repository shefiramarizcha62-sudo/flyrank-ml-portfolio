## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Research Question
Which content pages should be prioritized for review based on observable search and content performance signals?

### Decision
Determine which pages are the highest-priority candidates for content review and refresh.

### Action
Provide a ranked list of pages with supporting reason codes so that a content or SEO reviewer can decide which pages to review first.

### Cost of a Wrong Recommendation
If the model recommends pages that do not actually need review (false positives), reviewers waste time on low-value pages. If it misses pages that truly need attention (false negatives), valuable opportunities to improve search performance may be overlooked. Therefore, the system should support prioritization rather than replace human judgment.
